```python
import numpy as np

x = np.array([0, 1, 2, 45, 32, 67, 34, 79, 23, 12])

print(np.ndim(x))        # Number of dimensions
print(np.size(x))        # Size of array
print(np.shape(x))       # Shape of array
print(np.sum(x))         # Sum of elements
print(np.mean(x))        # Mean of elements
print(np.sort(x))        # Sorted array
print(np.sin(np.pi/3))   # Sine value
print(np.cos(0))         # Cosine value
```
