```python
import pandas as pd 
import numpy as np

df=pd.read_csv("data.csv")
data=df["Calories"]


mean=np.mean(data)
median=np.median(data)
mode=data.mode()[0]
variance=np.var(data)
std_deviation=np.std(data)

print("Mean:",mean)
print("Median:",median)
print("Mode:",mode)
print("Variance:",variance)
print("Standard Deviation:",std_deviation)
```
