```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

iris=load_iris()
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
x=df['sepal length (cm)']
y=df['petal length (cm)']
print(df.head())


```

```python
model=LinearRegression()
x=x.values.reshape(-1,1)
y=y.values.reshape(-1,1)
model.fit(x,y)
y_pred=model.predict(x)
```

```python
plt.scatter(x[:, 0], y, color='blue', label='Actual Data')
plt.plot(x[:, 0], y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()

# Convert continuous predictions to discrete class labels for confusion matrix
y_class = (y > y.mean()).astype(int)
y_pred_class = (y_pred > y.mean()).astype(int)
```

```python
# use discrete class labels for confusion matrix (not continuous features)

cm = confusion_matrix(y_class.ravel(), y_pred_class.ravel())
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show() 
```
